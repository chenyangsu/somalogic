```python
import pandas as pd
import os

ROOT_DIR = os.getcwd()  # .../somalogic
file = os.path.join(ROOT_DIR, 'data', 'Changes_severity.csv')
```

```python
df = pd.read_csv(file)
df
```

### Group by Patient ID and then sort date sampling in ascending order

```python
df_sorted = df.groupby (["anonymized_patient_id"], as_index=False). apply (lambda x: x.sort_values(["date_sampling"]))
df_sorted
```

```python
# Get subset rows with a change in A3_v2 case to control using 'change_A3' 
df_change = df_sorted[df_sorted['change_A3'] > 1]  # 1 = no change in A3_v2 case-control, 2 = change
df_change
```

```python
# get difference in change in each group
df_change['A3_v2_diff'] = df_change.groupby(["anonymized_patient_id"])['A3_v2'].diff()
```

```python
df_change
```

```python
df_filtered = df_change[df_change['A3_v2_diff'] > 0]  # only keep patients where change is greater than 1 
```

```python
df_filtered
```

```python
patients = df_filtered['anonymized_patient_id'].tolist()  # get the patients in list 
patients
```

```python
df_final = df_sorted[df_sorted['anonymized_patient_id'].isin(patients)]  # use patients list to get all samples of those patients
df_final
```

### Add days_symptom_update3 to last column

```python
file = os.path.join(ROOT_DIR, 'data', 'Proteomic_new_severity.csv')

proteomic_severity = pd.read_csv(file)
proteomic_severity
```

```python
to_merge = proteomic_severity[['SubjectID', 'Days_symptom_update3']]
to_merge
```

```python
df_final_final = pd.merge(df_final, to_merge, on='SubjectID')
df_final_final
```

```python
df_final_final.to_csv('A3_v2_control_to_case.csv', index=False)
```
